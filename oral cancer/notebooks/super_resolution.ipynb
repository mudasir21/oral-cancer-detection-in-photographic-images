Super Resolution Notebook

```python
#super resolution of preocessed_dataset in processing notebook
import cv2
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img_path = 'D:\Development\machine_learning\PROJECTS\oral_cancer_project\datasets\images\kag_1\oral_cancer_data\cancer'
```

ESPCN MODEL

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
 
path = "./models/ESPCN_x 2.pb"
 
sr.readModel(path)
 
sr.setModel("espcn",2)

img = img_path + '/1.jpg'

pil_image = Image.open(img)
res_img = np.array(pil_image)
result = sr.upsample(res_img)

print(result)

processed_image = Image.fromarray(result)
processed_image.save('processed_image.jpg')
processed_image.show()
```

EDSR MODEL

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
 
path = "./models/EDSR_x3.pb"
 
sr.readModel(path)
 
sr.setModel("edsr",3)

img = img_path + '/1.jpg'

pil_image = Image.open(img)
res_img = np.array(pil_image)
result = sr.upsample(res_img)

print(result)

processed_image = Image.fromarray(result)
processed_image.save('processed_image.jpg')
processed_image.show()

 
# Resized image
resized = cv2.resize(res_img, dsize=None, fx=3, fy=3)
 
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
# Original image
plt.imshow(res_img[:,:,::-1])
plt.subplot(1,3,2)
# SR upscaled
plt.imshow(result[:,:,::-1])
plt.subplot(1,3,3)
# OpenCV upscaled
plt.imshow(resized[:,:,::-1])
plt.show()
```
