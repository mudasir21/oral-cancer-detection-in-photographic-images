```python
import cv2
import os
import matplotlib.pyplot as plt 
from skimage.metrics import structural_similarity as compare_ssim
from skimage.metrics import peak_signal_noise_ratio as compare_psnr
```

```python
PATH = 'c:/Users/Groot/Code/oral_cancer/images'
```

```python
img = cv2.imread(PATH +'/oc_40_ckd.jpg')
new_img = cv2.resize(img, (512,512))
```

```python
img_des = PATH+'/new.jpg'
cv2.imwrite(img_des, new_img)
```

```python
imageA =  cv2.imread(PATH +'/new.jpg')
imageA =  cv2.resize(imageA, (512,512))

imageB =  cv2.imread(PATH +'/cropped.jpg')
imageB =  cv2.resize(imageB, (512,512))
```

```python

grayA = cv2.cvtColor(imageA, cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_BGR2GRAY)
```

```python
(score, diff) = compare_ssim(grayA, grayB, full=True)
print("SSIM: {}".format(score))
# print(diff)
```

```python
ratio = compare_psnr(imageA, imageB)
print("PSNR: {}".format(ratio))
```

```python
imageC = cv2.imread(PATH+'/oral_cancer_40_ckd.jpg')
```

```python
hist1 = cv2.calcHist([imageA],[0],None,[256],[0,255]) 
hist2 = cv2.calcHist([imageB],[0],None,[256],[0,256]) 
hist3 = cv2.calcHist([imageC],[0],None,[256],[0,256]) 
```

```python
hist1 = hist1/hist1.sum()
hist2 = hist2/hist2.sum()
hist3 = hist3/hist3.sum()
```

```python
hist4 = cv2.calcHist([imageA],[1],None,[256],[0,255]) 
hist5 = cv2.calcHist([imageB],[1],None,[256],[0,256]) 
hist6 = cv2.calcHist([imageC],[1],None,[256],[0,256]) 
```

```python
hist4 = hist4/hist4.sum()
hist5 = hist5/hist5.sum()
hist6 = hist6/hist6.sum()
```

```python
hist7 = cv2.calcHist([imageA],[2],None,[256],[0,255]) 
hist8 = cv2.calcHist([imageB],[2],None,[256],[0,256]) 
hist9 = cv2.calcHist([imageC],[2],None,[256],[0,256]) 
```

```python
hist7 = hist7/hist7.sum()
hist8 = hist8/hist8.sum()
hist9 = hist9/hist9.sum()
```

```python

plt.plot(hist3,label="Orgional")
plt.plot(hist2,label='Cropped and Resized')
plt.plot(hist1,label='Resized')
plt.legend()
plt.xlabel("Pixels")
plt.ylabel("Normalized Frequency")
plt.title('Histogram Channel R')
plt.show()
```

```python
plt.plot(hist6,label='Origional')
plt.plot(hist5,label='Cropped and Resized')
plt.plot(hist4,label='Resized')
plt.xlabel("Pixels")
plt.ylabel("Normalized Frequency")
plt.title('Histogram Channel G')
plt.legend()
plt.show()
```

```python
plt.plot(hist9,label="Origional")
plt.plot(hist7,label='Resized')
plt.plot(hist8,label= "Cropped and Resized")
plt.legend()
plt.xlabel("Pixels")
plt.ylabel("Normalized Frequency")
plt.title('Histogram Channel B')
```
